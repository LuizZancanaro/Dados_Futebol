##Importando os datasets

### Bibliotecas

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
import locale
```

###Importando os datasets

```python
db_goalscores_0 = pd.read_csv('https://raw.githubusercontent.com/LuizZancanaro/Dados_Futebol/main/goalscorers.csv', sep = ',')
db_results_0 = pd.read_csv('https://raw.githubusercontent.com/LuizZancanaro/Dados_Futebol/main/results.csv', sep = ',')
db_shootouts_0 = pd.read_csv('https://raw.githubusercontent.com/LuizZancanaro/Dados_Futebol/main/shootouts.csv', sep = ',')
```

###Coluna perdedores

Uma automação que indica quem perdeu a partida.

```python
db_shootouts_0['looser'] = " "

for i in range(len(db_shootouts_0['date'])):
  if db_shootouts_0['home_team'][i] == db_shootouts_0['winner'][i]:
    db_shootouts_0['looser'][i] = db_shootouts_0['away_team'][i]
  else:
    db_shootouts_0['looser'][i] = db_shootouts_0['home_team'][i]


db_shootouts_0.head()
```

###Vencedor em casa e Perdedor em casa

Uma automação que vai criar duas colunas no dataset para indicar quem foi o time que perdeu ou ganhou em casa.

```python
db_shootouts_0['home_winner'] = "_"
db_shootouts_0['home_looser'] = "_"


for i in range(len(db_shootouts_0['date'])):
  if db_shootouts_0['home_team'][i] == db_shootouts_0['winner'][i]:
    db_shootouts_0['home_winner'][i] = db_shootouts_0['home_team'][i]
  else:
    db_shootouts_0['home_looser'][i] = db_shootouts_0['home_team'][i]


db_shootouts_0.head()
```

##Convertendo as datas

Como vai ser criado um filtro para selecionar o intervalo de datas a ser estudado é necessário a conversão do dataset de str para date.

```python
for i in range(len(db_goalscores_0['date'])):
  db_goalscores_0['date'][i] = datetime.strptime(db_goalscores_0['date'][i], '%Y-%m-%d').date()
```

```python
for i in range(len(db_results_0['date'])):
  db_results_0['date'][i] = datetime.strptime(db_results_0['date'][i], '%Y-%m-%d').date()
```

```python
for i in range(len(db_shootouts_0['date'])):
  db_shootouts_0['date'][i] = datetime.strptime(db_shootouts_0['date'][i], '%Y-%m-%d').date()
```

## Criar um filtro

Esse filtro foi criado com o objetivo de fazer uma análise mais especifica de acordo com as preferências do cliente.


```python
data_max = max(max(db_goalscores['date']),   max(db_shootouts['date']), max(db_results['date']))


data_min = min(min(db_goalscores['date']),   min(db_shootouts['date']), min(db_results['date']))

print(f'A ultima partida registrada é {data_max}, a primeira partida registrada é {data_min}.')
```

```python
print(f'A pesquisa pode ser feita entre {data_min} e {data_max}!')
a_comeco = input('Digite o ano para o começo da pesquisa: ')

a_fim = input(f'Escolha um ano para o fim da pesquisa: ')


while a_comeco > a_fim:
  print(f"ANO INVÁLIDO")
  a_fim = input(f'Escolha um ano maior que {int(a_comeco)-1}: ')

a_comeco = a_comeco + '-01-01'
a_fim = a_fim + '-12-31'

date_comeco = datetime.strptime(a_comeco, '%Y-%m-%d').date()
date_fim = datetime.strptime(a_fim, '%Y-%m-%d').date()



```

##Aplicando o filtro


```python
db_goalscores = db_goalscores_0
db_results = db_results_0
db_shootouts = db_shootouts_0



db_goalscores = db_goalscores[(db_goalscores['date'] >= date_comeco) & (db_goalscores['date'] <= date_fim)]
db_results = db_results[(db_results['date'] >= date_comeco) & (db_results['date'] <= date_fim)]
db_shootouts = db_shootouts[(db_shootouts['date'] >= date_comeco) & (db_shootouts['date'] <= date_fim)]
```

###Maior número de vitórias

```python
vencedores = db_shootouts['winner'].unique()
vencedores = sorted(vencedores)
df_vencedores = db_shootouts['winner'].value_counts().to_frame()

print(f'No período estudado temos {len(vencedores)} times vencedores, o que mais venceu no período foi o {df_vencedores.index[0]}!')
df_vencedores.rename(columns={df_vencedores.columns[0]: 'Qtd_Vitórias'}).head(5)
```

###Maior número de derrotas

```python
perdedores = db_shootouts['looser'].unique()
perdedores = sorted(perdedores)
df_perdedores = db_shootouts['looser'].value_counts().to_frame()

print(f'No período estudado temos {len(perdedores)} times perdedores, o que mais perdeu no período foi o {df_perdedores.index[0]}!')
df_perdedores.rename(columns={df_perdedores.columns[0]: 'Qtd_derrotas'}).head(5)

```

###Aproprimorando a tabela de resultados

O objetivo vai ser acrescentar na tabela quantas vezes cada time ganhou em casa, perdeu em casa, ganhou fora de casa e perdeu fora de casa.





```python
##vitorias = vitorias em casa + vitorias fora de casa (V = VC + VF)
##analogamente derrotas = derrotas em casa + derrotas fora de casa (D = DC + DF)

db_shootouts.head()
```

```python
vencedores_casa = db_shootouts['home_winner'].unique()
vencedores_casa = sorted(vencedores_casa)
df_vencedores_casa = db_shootouts['home_winner'].value_counts().to_frame()
a = df_vencedores_casa.index[0]
df_vencedores_casa = df_vencedores_casa.drop(a)
print(f'No período estudado temos {len(vencedores_casa)} times vencedores em casa, o que mais venceu em casa no período foi o {df_vencedores_casa.index[0]}!')

df_vencedores_casa.rename(columns={df_vencedores_casa.columns[0]: 'Qtd_Vitórias_casa'}).head(5)
```

```python
perdedores_casa = db_shootouts['home_looser'].unique()
perdedores_casa = sorted(perdedores_casa)
df_perdedores_casa = db_shootouts['home_looser'].value_counts().to_frame()
a = df_perdedores_casa.index[0]
df_perdedores_casa = df_perdedores_casa.drop(a)
print(f'No período estudado temos {len(perdedores_casa)} times vencedores em casa, o que mais venceu em casa no período foi o {df_perdedores_casa.index[0]}!')

df_perdedores_casa.rename(columns={df_perdedores_casa.columns[0]: 'Qtd_Derrotas_casa'}).head(5)
```

###Juntando os datasets

Assim teremos um panorama geral da situação de cada time.


```python
#pd.merge(left2, right2, left_index = True, right_index = True, how = "inner, outer, left, right")

resultados_1 = pd.merge(df_vencedores, df_perdedores, left_index = True, right_index = True, how = 'outer')
resultados_2 = pd.merge(df_vencedores_casa, df_perdedores_casa, left_index = True, right_index = True, how = 'outer')
resultados_3 = pd.merge(resultados_1, resultados_2, left_index = True, right_index = True, how = 'outer')
resultados_3.fillna(0)
```

###Derrotas em casa e vitórias fora de casa.

```python
resultados_3['away_looser'] = resultados_3['looser'] - resultados_3['home_looser']
resultados_3['away_winner'] = resultados_3['winner'] - resultados_3['home_winner']

resultados_3 = resultados_3.fillna(0)
resultados_3 = resultados_3[['winner', 'home_winner', 'away_winner', 'looser', 'home_looser', 'away_looser']]
resultados_3
```
